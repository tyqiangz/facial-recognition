# Facial Recognition

Using Python package from https://github.com/ageitgey/face_recognition

Given an image, find all the locations (pixel-wise) of the faces, if any.

Then check if the faces tallies with any of the known faces.

```python
import face_recognition
from matplotlib import pyplot as plt
```

```python
known_image = face_recognition.load_image_file("../images/joe_biden.jfif")
unknown_image = face_recognition.load_image_file("../images/unknown_0.jpg")
```

```python
plt.figure(figsize=(20,10))
plt.imshow(known_image)
plt.show()
```

```python
plt.figure(figsize=(20,10))
plt.imshow(unknown_image)
plt.show()
```

```python
biden_encoding = face_recognition.face_encodings(known_image)[0]
unknown_encoding = face_recognition.face_encodings(unknown_image)[0]
```

```python
results = face_recognition.compare_faces([biden_encoding], unknown_encoding)
results
```
