# Facial Detection

Using Python package from https://github.com/ageitgey/face_recognition

Given an image, find all the locations (pixel-wise) of the faces, if any.

```python
import face_recognition
from matplotlib import pyplot as plt
```

Load the image containing faces

```python
image = face_recognition.load_image_file("../images/politician_baey_yam_keng/00012.jpg")
```

Plot the image

```python
plt.figure(figsize = (20,10))
plt.imshow(image)
plt.show()
```

Begin facial recognition

```python
face_locations = face_recognition.face_locations(image)

print("I found {} face(s) in this photograph.".format(len(face_locations)))
```

Plot all the faces

```python
for face_location in face_locations:

    # Print the location of each face in this image
    top, right, bottom, left = face_location
    print("A face is located at pixel location Top: {}, Left: {}, Bottom: {}, Right: {}".format(top, left, bottom, right))

    # You can access the actual face itself like this:
    face_image = image[top:bottom, left:right]
    plt.imshow(face_image)
    plt.show()
```
